## Question 1
How has AI changed globaly in the past 7 years?

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
```

```python
globalAI = pd.read_csv('AI_index_db.csv')

countries_top10 = globalAI['Country'].iloc[0:10]
countries_bottom10 = globalAI['Country'].iloc[-10:] 
talent_top10 = globalAI['Talent'].iloc[0:10].astype(float) 
talent_bottom10 = globalAI.iloc[-10:, 1].astype(float) 
Infrastrucure_top10 = globalAI.iloc[0:10, 2].astype(float)  
Infrastructure_bottom10 = globalAI.iloc[-10:, 2].astype(float)
Operating_bottom10 = globalAI.iloc[-10:, 3].astype(float) 
Operating_top10 = globalAI.iloc[0:10, 3].astype(float)
Research_top10 = globalAI.iloc[0:10, 4].astype(float) 
Research_bottom10 = globalAI.iloc[-10:, 4].astype(float) 
Development_top10 = globalAI.iloc[0:10, 5].astype(float)  
Development_bottom10 = globalAI.iloc[-10:, 5].astype(float)
Government_bottom10 = globalAI.iloc[-10:, 6].astype(float) 
Government_top10 = globalAI.iloc[0:10, 6].astype(float)
Commercial_top10 = globalAI.iloc[0:10, 7].astype(float) 
Commercial_bottom10 = globalAI.iloc[-10:, 7].astype(float) 



plt.figure(figsize=(12, 8))
plt.plot(countries_top10, talent_top10, label='Talent (Top 10 Countries)', marker='o')
plt.plot(countries_bottom10, talent_bottom10, label='Talent (Bottom 10 Countries)', marker='x')
plt.plot(countries_top10, Infrastrucure_top10, label='Infrastructure (Top 10 Countries)', marker ='v')
plt.plot(countries_bottom10, Infrastructure_bottom10, label='Infrastructure (Bottom 10 Countries)', marker='*')
plt.plot(countries_top10, Operating_top10, label='Operating Environment (Top 10 Countries)', marker='o')
plt.plot(countries_bottom10, Operating_bottom10, label='Operating Environment (Bottom 10 Countries)', marker='x')
plt.plot(countries_top10, Research_top10, label='Research (Top 10 Countries)', marker ='v')
plt.plot(countries_bottom10, Research_bottom10, label='Research (Bottom 10 Countries)', marker='*')
plt.plot(countries_top10, Development_top10, label='Development (Top 10 Countries)', marker ='v')
plt.plot(countries_bottom10, Development_bottom10, label='Development (Bottom 10 Countries)', marker='*')
plt.plot(countries_top10, Government_top10, label='Government (Top 10 Countries)', marker='o')
plt.plot(countries_bottom10, Government_bottom10, label='Government (Bottom 10 Countries)', marker='x')
plt.plot(countries_top10, Commercial_top10, label='Commercial (Top 10 Countries)', marker ='v')
plt.plot(countries_bottom10, Commercial_bottom10, label='Commercial (Bottom 10 Countries)', marker='*')
plt.title("Top and Bottom 10 Countries based on their involvment with AI", fontsize=16)
plt.xlabel("Countries", fontsize=12)
plt.ylabel("Scores", fontsize=12)
plt.legend(fontsize=6)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(
    ticks=range(len(countries_top10) + len(countries_bottom10)),
    labels=list(countries_top10) + list(countries_bottom10),  
    rotation=45,
    ha='right' 
)
plt.tight_layout()
plt.show()
```

## Question 3
How has the job market been affected by the increase in AI?

Dataset: https://www.kaggle.com/datasets/muhammadroshaanriaz/the-rise-of-artificial-intelligence

```python
df = pd.DataFrame({
    'years': list(range(2018, 2025)),
    'jobelimAI': Rise_AI.iloc[1:9,7],
    'jobcreateAI': Rise_AI.iloc[1:9, 8],
    'netjobloss': Rise_AI.iloc[1:9, 9]
})

df = df.astype('float')

plt.figure(figsize=(10, 6))
plt.plot(df['years'], df['jobelimAI'], label = 'Estimated Jobs Eliminated by AI')
plt.plot(df['years'], df['jobcreateAI'], label = 'Estimated Jobs Created by AI')
plt.plot(df['years'], df['netjobloss'], label = 'Net Job Loss in the US')
plt.title("AI in the Job Market", fontsize=16)
plt.xlabel("Years", fontsize=12)
plt.ylabel("Percent", fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.6)
```
